### Iterate through the regions and repeat the calculating MA process
#### - Series to Data Frame:
1. df["column"].to_frame()
2. df[["column"]]

#### - Add [:8] to prevent memory explosion!!
- Probably due to type dupicates, e.g. Two rows have identical date(index), since they only differ in type, i.e. conventional/organic

```python
graph_df = pd.DataFrame()    # Create a dataframe

for region in df['region'].unique()[:8]:
    print(region)
    region_df = df.copy()[df['region'] == region]
    region_df.set_index("Date", inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f"{region}price_20MA"] = region_df['AveragePrice'].rolling(20).mean()
    
    if graph_df.empty:
        graph_df = region_df[f"{region}price_20MA"].to_frame()
#         graph_df = region_df["price_20MA"]
#         print(type(graph_df))   -->   This returns a series, I want a data frame so I can continue to add columns
#         graph_df = region_df[["price_20MA"]] can also work
    else:
        graph_df = graph_df.join(region_df[f"{region}price_20MA"])
    
```

### Restart from the beginning to fix memory issue:

```python
import pandas as pd
df = pd.read_csv("avocado.csv")
df = df.copy()[df['type'] == "organic"]   # Add .copy() in between to get rid of a warning
df['Date'] = pd.to_datetime(df['Date'])   # Convert "Date" column to Datetime objects

df.sort_values(by="Date", ascending=True, inplace=True)    # by="Date""Date"
df.head(5)
```

#### Paste our previous graph_df code @ln26 again

```python
graph_df = pd.DataFrame()    # Create a dataframe

for region in df['region'].unique():
    print(region)
    region_df = df.copy()[df['region'] == region]
    region_df.set_index("Date", inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f"{region}price_20MA"] = region_df['AveragePrice'].rolling(20).mean()

    if graph_df.empty:
        graph_df = region_df[f"{region}price_20MA"].to_frame()
#         graph_df = region_df["price_20MA"]
#         print(type(graph_df))   -->   This returns a series, I want a data frame so I can continue to add columns
#         graph_df = region_df[["price_20MA"]] can also work
    else:
        graph_df = graph_df.join(region_df[f"{region}price_20MA"])
```
